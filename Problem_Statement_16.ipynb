```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from scipy import stats
```

```python
# Load the dataset
df = pd.read_csv('adult_dataset.csv')

# m. Data Cleaning
# Remove NA values
df = df.dropna()
```

```python
df
```

```python
# Replace '?' with NaN and drop rows with NaN
df = df.replace('?', np.nan).dropna()
```

```python
df
```

```python
# Remove negative values in numeric columns
numeric_cols = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
df = df[(df[numeric_cols] >= 0).all(axis=1)]
```

```python
numeric_cols
```

```python
df
```

```python
# n. Error Correcting (Outlier detection and removal)
# Using Z-score for outlier detection in numeric columns
z_scores = stats.zscore(df[numeric_cols])
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df = df[filtered_entries]
```

```python
df
```

```python
# o. Data Transformation
# Encode categorical variables
categorical_cols = ['workclass', 'education', 'marital-status', 'occupation', 
                    'relationship', 'race', 'gender', 'native-country']
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
# Encode target variable (income)
df['income'] = df['income'].map({'>50K': 1, '<=50K': 0})
```

```python
# Separate features and target
X = df.drop(['income', 'fnlwgt'], axis=1)  # Dropping fnlwgt as it's not meaningful for prediction
y = df['income']
```

```python
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# p. Build Data Models
# Logistic Regression
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_predictions)
```

```python
# Naive Bayes
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
nb_predictions = nb_model.predict(X_test)
nb_accuracy = accuracy_score(y_test, nb_predictions)
```

```python

# Print results
print(f"Logistic Regression Accuracy: {lr_accuracy:.4f}")
print(f"Naive Bayes Accuracy: {nb_accuracy:.4f}")
```

```python
# Compare models
if lr_accuracy > nb_accuracy:
    print("Logistic Regression performs better for income prediction")
elif nb_accuracy > lr_accuracy:
    print("Naive Bayes performs better for income prediction")
else:
    print("Both models perform equally well for income prediction")
```
