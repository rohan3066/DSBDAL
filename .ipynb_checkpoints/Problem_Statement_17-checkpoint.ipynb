```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('heart.csv')

# Set style for better visualization
print(plt.style.available)

```

```python
plt.style.use('ggplot')
```

```python
fig = plt.figure(figsize=(20, 15))
```

```python
df
```

```python
# a. Histograms
# Objective: Show the distribution of continuous variables (age, trestbps, chol, thalach)
plt.subplot(3, 2, 1)
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(3, 2, 2)
plt.hist(df['chol'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Histogram of Cholesterol')
plt.xlabel('Cholesterol (mg/dl)')
plt.ylabel('Frequency')
```

```python
# b. Dot Plots (using strip plots)
# Objective: Display individual data points for categorical vs continuous variables
plt.subplot(3, 2, 3)
sns.stripplot(x='target', y='thalach', data=df, color='purple', alpha=0.6)
plt.title('Dot Plot: Max Heart Rate by Target')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Max Heart Rate (thalach)')
```

```python
# c. Bar Plots
# Objective: Show the count of categorical variables (sex, cp, target)
plt.subplot(3, 2, 4)
sns.countplot(x='sex', hue='target', data=df, palette='Set2')
plt.title('Bar Plot: Heart Disease by Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
```

```python
# d. Line Charts
# Objective: Show trends of continuous variables across age groups
age_groups = pd.cut(df['age'], bins=5)
avg_chol = df.groupby(age_groups)['chol'].mean()
plt.subplot(3, 2, 5)
avg_chol.plot(kind='line', marker='o', color='coral')
plt.title('Line Chart: Average Cholesterol by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Cholesterol')
plt.xticks(rotation=45)
```

```python
# e. Combined Box Plot with Histogram and Scatter
# Objective: Combine distribution (histogram), spread (box plot), and individual points (scatter)
plt.subplot(3, 2, 6)
# Create a secondary axis for the histogram
ax1 = plt.gca()
ax2 = ax1.twinx()

# Box plot
sns.boxplot(y='thalach', data=df, color='lightblue', width=0.2, ax=ax1)
# Histogram
sns.histplot(y='thalach', data=df, color='orange', alpha=0.5, stat='density', bins=20, ax=ax2)
# Scatter plot
sns.stripplot(y='thalach', data=df, color='red', alpha=0.3, size=3, ax=ax1)

# Adjust axes labels and title
ax1.set_ylabel('Max Heart Rate (thalach)')
ax2.set_ylabel('Density')
plt.title('Combined: Box Plot, Histogram, and Scatter of Max Heart Rate')

# Adjust layout and save
plt.tight_layout()
plt.savefig('heart_disease_visualizations.png')
```
