```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('hepatitis_csv.csv')
```

```python
df
```

```python
df.replace('',np.nan,inplace=True)
```

```python
df
```

```python
numeric_cols = ['age', 'bilirubin', 'alk_phosphate', 'sgot', 'albumin', 'protime']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df
```

```python
male_subset=df[df['sex']=='male']
female_subset=df[df['sex']=='female']
```

```python
male_subset
```

```python
female_subset
```

```python
merged_df=pd.concat([male_subset,female_subset],ignore_index=True)
```

```python
merged_df
```

```python
sorted_df = df.sort_values(by=['age', 'sgot', 'protime'], na_position='last')
```

```python
sorted_df
```

```python
transpose_df=df.transpose()
```

```python
transpose_df
```

```python
melted_df=pd.melt(
    df,
    id_vars=['age','sex','class'],
    value_vars=['bilirubin', 'alk_phosphate', 'sgot', 'albumin', 'protime'],
    var_name='Measurment',
    value_name='Value'
)
```

```python
melted_df
```

```python
wide_df = melted_df.pivot_table(
    index=['sex', 'class'],
    columns='Measurment',
    values='Value',
  # Use mean to handle potential duplicates
).reset_index()
```
