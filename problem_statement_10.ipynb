```python
df['ID']=df.index;
melted_df=pd.melt(
    df,
    id_vars=['ID','variety'],
    value_vars=['sepal.length','sepal.width','petal.length','petal.width'],
    var_name='measurement',
    value_name='value'
)
```

```python
melted_df
```

```python
wide_df=melted_df.pivot(
    index=['variety','ID'],
    columns='measurement',
    values='value'
).reset_index()
```
