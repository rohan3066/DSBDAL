```python
df['ID']=df.index;
melted_df=pd.melt(
    df,
    id_vars=['ID','age','gender','race'],
    value_vars=['capital-gain','capital-loss','hours-per-week'],
    var_name='Metric',
    value_name='Value'
)
```

```python
melted_df
```

```python
wide_df=melted_df.pivot(
    index=['age','gender','race','ID'],
    columns='Metric',
    values='Value'
).reset_index();
```
